## Approaches
1. Use Backtrader + strategies. Can optimize hyperparameters by values/statistics that backtrader outputs, such as account value, ratios, etc.
2. Use labels on each tick of trading data, i.e. ternary classification problem with {-1, 1} indicating price movement. Use these labels to optimize strategies. Final execution of trades will be done based on probability thresholds for the labels, generated by the model.
3. Attempt to forecast price

1 optimizes based on account value, while 2 optimizes based on the price data.

### Idea 1
1. Use machine learning techniques on dataset to produce weights, etc. Thus, will optimize on either labels or price forecast errors. Use with training/cross-validation set.
2. Use produced weights from Step 1 in Backtrader to calculate account value, etc. in "real-world" conditions. Use with test set.
